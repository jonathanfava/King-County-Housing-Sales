## 3. Q4 - Scatterplots

```python
# create a scatterplot for price and living squarefotage

sns.lmplot(x = 'price', y = 'sqft_living', data = df)
```

### You can see that there is a upward trend line showing that as living sqaure footage increases so does the price. With that being said there are also some outliers that sold for low prices even though they have a lot of sqaure footage.

```python
# create a scatterplot for price and bathrooms

sns.lmplot(x = 'price', y = 'bathrooms', data = df)
```

### There is an upward trend in price as the number of bathrooms increases. Which can be assumed that homes with more bathrooms are inturn worth more.

## 5. Q6 Categorical Plot

```python
# create a categorical plot comparing price by living square footage with zipcode as the categorical variable

sns.set(style="ticks")
zip_price_lvsqft = sns.catplot(x="sqft_living", y="price", hue="zipcode", data=df2)
```

## Since there are many zipcodes this categorical plot makes it difficult to interpret, but it is clear that as livable square footage increases so does the price and each zipcode has a wide range of home sizes and prices.
